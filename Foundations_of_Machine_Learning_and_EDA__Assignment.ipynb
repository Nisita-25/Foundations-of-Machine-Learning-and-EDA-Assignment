# Question 2: Explain overfitting and underfitting in ML. How can you detect and prevent them?
- 1. Overfitting

Definition:
Overfitting happens when a machine learning model learns the training data too well, including noise and unnecessary details. As a result, it performs very well on training data but poorly on new, unseen data.

How to Detect Overfitting:

Training accuracy is very high, but testing/validation accuracy is low.

Large gap between training and test errors.

Validation loss increases after some epochs (in neural networks).

How to Prevent Overfitting:

Cross-validation: Use k-fold cross-validation to ensure the model performs well on different data splits.

Regularization: Apply L1/L2 regularization, dropout (for DL), or early stopping.

Simplify the model: Reduce model complexity (e.g., use fewer layers/parameters).

Use more data: Increase training data or apply data augmentation.

2. Underfitting

Definition:
Underfitting occurs when a model is too simple to learn the underlying pattern in the data. It performs poorly on both training and test data.

How to Detect Underfitting:

Both training and testing accuracy are low.

High bias: model makes overly simple assumptions.

Training loss remains high.

How to Prevent Underfitting:

Increase model complexity: Use deeper networks, more features, or more advanced algorithms.

Train longer: More epochs or lower learning rate.

Feature engineering: Add meaningful features or reduce noise.

Decrease regularization: Reduce L1/L2 penalties if they are too strong.

# Question 3:How would you handle missing values in a dataset? Explain at least three methods with examples.
- Handling Missing Values in a Dataset

Missing values are common in real-world datasets, and handling them properly is important for building accurate machine learning models. Here are three commonly used methods:

1. Deletion Methods

a. Listwise Deletion (Remove entire row)
If a row contains missing data, the whole row is removed.

Example:
Dataset:
Age = [25, 30, NaN, 40]
Income = [50k, 60k, NaN, 80k]
â†’ Remove row with the missing values.

When to Use:

When the dataset is large.

Missing values are very few.

Risk:

Loss of useful data if missing values are many.

2. Imputation Using Mean/Median

Replace missing values with statistical measures of the column.

a. Mean Imputation
Replace missing value with the mean of the column.

Example:
Ages = [20, 25, NaN, 35]
Mean = (20+25+35)/3 = 26.67
Missing value becomes 26.67.

Best For:

Numerical data with no outliers.

b. Median Imputation
Use the median instead of mean.

Example:
Income = [30k, 35k, NaN, 500k]
Median = 35k (better than mean because of outlier 500k)

Best For:

Numerical data with outliers.

3. Predictive Modeling (Advanced Imputation)

Use machine learning models to predict the missing value.

How it Works:

Treat the missing value as a target variable.

Use remaining columns as features.

Apply algorithms such as KNN, Linear Regression, or Decision Trees to estimate the missing value.

Example:
Missing "Salary" value can be predicted using features like:
Education + Age + Experience â†’ Predicted Salary

Advantages:

More accurate than simple mean/median.

Preserves relationships between variables.

# Question 4:What is an imbalanced dataset? Describe two techniques to handle it (theoretical + practical).
- Imbalanced Dataset
An imbalanced dataset is a dataset where the number of samples in one class is much larger than the other.
This causes the model to become biased toward the majority class and perform poorly on the minority class.

Example:
Fraud detection dataset:

Legit transactions: 98%

Fraud transactions: 2%
The model may predict everything as â€œlegitâ€ and still get 98% accuracy, but it fails in detecting fraud.

Techniques to Handle Imbalanced Data
1. Random Oversampling
Theory:

This method increases the size of the minority class by randomly duplicating its samples until the dataset becomes balanced.

Practical Example:

Dataset:

Class 0: 900 samples

Class 1: 100 samples (minority)

Oversample Class 1 â†’ duplicate samples â†’ make it 900 samples.
New dataset:

Class 0: 900

Class 1: 900

Advantages:

Simple and effective

Useful when data is limited

Disadvantages:

May lead to overfitting because the model sees exact duplicates

2. Random Undersampling
Theory:

This method reduces the majority class by randomly removing samples to match the minority class.

Practical Example:

Dataset:

Class 0: 900

Class 1: 100

Undersample Class 0 â†’ remove 800 samples â†’ make it 100 samples.
New dataset:

Class 0: 100

Class 1: 100

Advantages:

Fast and reduces training time

Disadvantages:

Loss of useful data

May reduce model performance

3. SMOTE (Synthetic Minority Oversampling Technique)
Theory:

SMOTE creates synthetic examples for the minority class instead of duplicating existing ones.
It generates new samples by interpolating between a minority sample and its nearest neighbors.

Practical Example:

If Class 1 has a data point X = (2,4) and a neighbor Y = (3,5),
SMOTE may generate:
New point = (2.5, 4.5)

Advantages:

Reduces overfitting

Produces more diverse training examples

Disadvantages:

May create noisy or unrealistic samples

Needs careful parameter tuning

# Question 6: Compare Label Encoding and One-Hot Encoding. When would you prefer one over the other?
-1. Label Encoding
What it is:

Label Encoding converts categorical text values into numerical labels (0, 1, 2, 3, â€¦).

Example:

Fruit = {Apple, Banana, Orange}
â†’ Apple = 0, Banana = 1, Orange = 2

Advantages:

Simple and fast

Works well for ordinal data (data with order)

Disadvantages:

Introduces false numerical order for non-ordinal categories
(model may think Orange > Banana > Apple)

Can mislead algorithms like Linear Regression, KNN, SVM

Best Use Case:

Use Label Encoding for ordinal categories such as:

Size: Small < Medium < Large

Education Level: High school < Bachelor < Master < PhD

2. One-Hot Encoding
What it is:

One-Hot Encoding converts each categorical value into a separate binary column (0 or 1).

Example:

Fruit = {Apple, Banana, Orange}

Becomes:

Apple	Banana	Orange
1	0	0
0	1	0
0	0	1
Advantages:

No false ordering

Best for algorithms that rely on distance or linear relationships

Disadvantages:

Increases the number of columns (curse of dimensionality)

Not ideal when categories are very large (e.g., 10,000 items)

Best Use Case:

Use One-Hot Encoding for nominal data (no order), such as:

Country (India, USA, UK)

Color (Red, Blue, Green)

Job roles, cities, product categories

When to Prefer Which?
âœ”ï¸ Use Label Encoding when:

The categories have a natural order
(Low < Medium < High, Beginner < Intermediate < Expert)

Using tree-based models
(Decision Trees, Random Forest, XGBoost), because they are not affected by numerical ordering.

âœ”ï¸ Use One-Hot Encoding when:

The categories have no order (nominal)

Using distance-based or linear models, such as:

Linear Regression

Logistic Regression

KNN

SVM

Neural Networks
Because these models get confused by numeric labels.
